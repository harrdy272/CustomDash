#### importing basic libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#### Getting the dataset

```python
dataset = pd.read_csv('practice.csv')
```

#### Cropping dataset for meaningful values

```python
dataset = dataset.iloc[:,2:36].values
```

#### Visualizing the data

```python
dataset = pd.DataFrame(dataset)
dataset
```

#### Masking columns of dataset to apply Encoder to Categorical data

```python
categorical_feature_mask = dataset.dtypes==object
categorical_cols = dataset.columns[categorical_feature_mask].tolist()
```

#### Handling categorical data using LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
dataset[categorical_cols] = dataset[categorical_cols].apply(lambda col: label.fit_transform(col))
dataset
```

#### Making dependent and independent variables

```python
X = dataset.iloc[:, dataset.columns != 14].values
y = dataset.iloc[:,14].values
```

#### Making training and test data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.1,random_state=0)
```

#### Scaling data using StandardScaler

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

#### Classification of data using RandomForest Classification

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators=20,criterion='entropy',random_state=0)
classifier.fit(X_train,y_train)
```

#### Predicting test values

```python
y_pred = classifier.predict(X_test)
```

#### Comparing Original y_test with y_pred

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
```

#### Predicting Accuracy

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
